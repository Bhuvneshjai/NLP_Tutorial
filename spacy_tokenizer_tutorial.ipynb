```python
import spacy
```

Create blank language object and tokenize words in a sentence

```python
nlp = spacy.blank("en")
doc = nlp("Dr. Strange loves pav bhaji of mumbai as it costs only 2$ per plate.")

for token in doc:
    print(token)
```

Creating blank language object gives a tokenizer and an empty pipeline. We will look more into language pipelines in next tutorial

#### Collecting email ids of students from students information sheet

```python
with open("students.txt") as f:
    text = f.readlines()
text
```

```python
text = " ".join(text)
text
```

```python
doc = nlp(text)
emails = []
for token in doc:
    if token.like_email:
        emails.append(token)
emails
```

#### Customizing tokenizer

```python
from spacy.symbols import ORTH

nlp = spacy.blank("en")
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python
nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"},
    {ORTH: "me"},
])

doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

#### Sentence Tokenization or Segmentation

```python
nlp.pipeline
```

```python
nlp.add_pipe('sentencizer')
```

```python
doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
for sentence in doc.sents:
    print(sentence)
```
